```python
import os
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
import seaborn as sns
```

```python
path_to_csv = os.path.join("..", "data", "athlete_events.csv")

athletes = pd.read_csv(path_to_csv)
```

```python
athletes.head()
```

```python
athletes["Medal"].replace(np.NaN, "None", inplace=True)
```

```python
athletes.dropna(how="any", inplace=True)
```

```python
athletes.groupby("Sport")["ID"].count()
```

```python
athletes.loc[athletes["Sport"] == "Cycling"].sort_values(by="Weight")
```

### To make dataset smaller, lets only look at a specific Sport 

```python
athletes_basketball = athletes.loc[athletes["Sport"] == "Basketball"]
```

## To predict Athletes progress, we need:
- Sex (get_dummies)
- Age
- Height
- Weight
- Year
- Sport (get_dummies)


```python
athletes_build = athletes_basketball.drop(columns=["ID", "Name", "Team", "NOC", "Games", "Season", "City", "Sport", "Medal"])
athletes_build.info()
```

```python
# Predict Age of each athlete
X = athletes_build[["Sex", "Age", "Weight", "Year", "Event"]]
y = athletes_build["Height"].values.reshape(-1, 1)
```

```python
# Get Dummies for Sex and Sport
n = pd.get_dummies(athletes_build.Sex)
X = pd.concat([X,n], axis=1)
m = pd.get_dummies(athletes_build.Event)
X = pd.concat([X,m], axis=1)
X.drop(["Sex", "Event"], inplace=True, axis=1)
X.head()
```

### Run through Linear Regression model to predict Age/Height/Weight

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
print("Linear Regression R squared: %.4f" % model.score(X_test, y_test))
```

### Height and weight predict fairly decently, but Age was pretty off. Height and Weight are sitting around 70%-80%

## The sklearn diagram says to use SGDRegressor

```python
from sklearn.linear_model import SGDRegressor
```

```python
from sklearn.preprocessing import StandardScaler
X_scaler = StandardScaler().fit(X_train)
y_scaler = StandardScaler().fit(y_train)


X_train_scaled = X_scaler.transform(X_train)
X_test_scaled = X_scaler.transform(X_test)
y_train_scaled = y_scaler.transform(y_train)
y_test_scaled = y_scaler.transform(y_test)
```

```python
N = len(X_train_scaled)

SGDmodelScaled = SGDRegressor(max_iter = np.ceil(10**6 / N), tol=1e-3)
SGDmodelScaled.fit(X_train_scaled, y_train_scaled)
```

```python
predictions = SGDmodelScaled.predict(X_test_scaled)
plt.scatter(SGDmodelScaled.predict(X_train_scaled), SGDmodelScaled.predict(X_train_scaled) - np.reshape(y_train_scaled, len(y_train_scaled)), c="blue", label="Training Data")
plt.scatter(SGDmodelScaled.predict(X_test_scaled), SGDmodelScaled.predict(X_test_scaled) - np.reshape(y_test_scaled, len(y_test_scaled)), c="orange", label="Testing Data")
plt.legend()
plt.hlines(y=0, xmin=y_test_scaled.min(), xmax=y_test_scaled.max())
plt.title("Residual Plot")
plt.show()
```

```python
# Used X_test_scaled, y_test_scaled, and model.predict(X_test_scaled) to calculate MSE and R2

### BEGIN SOLUTION
from sklearn.metrics import mean_squared_error

MSE = mean_squared_error(y_test_scaled, predictions)
r2 = SGDmodelScaled.score(X_test_scaled, np.reshape(y_test_scaled, len(y_test_scaled)))
### END SOLUTION

print(f"MSE: {MSE}, R2: {r2}")
```

```python
print("Linear Regression R squared: %.4f" % SGDmodelScaled.score(X_test_scaled, np.reshape(y_test_scaled, len(y_test_scaled))))
```
