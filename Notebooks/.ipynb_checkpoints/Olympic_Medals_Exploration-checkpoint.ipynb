# Olympic Medal Winners Exploration

```python
import os
import numpy as np
import pandas as pd
```

```python
path_to_csv = os.path.join("..", "data", "athlete_events.csv")

athletes = pd.read_csv(path_to_csv)
```

```python
athletes.head()
```

## Lets see how many medals each country has... is it accurate?

```python
# Create Groupby object for Team
country_group = athletes.groupby(by="Team")

# How many medals has each country won?
country_medal_count = country_group["Medal"].count()
```

```python
country_medal_count.sort_values(ascending=False)
```

#### Not Entirely sure if this is an accurate metric. Number seems about double of what wikipedia suggests. 
https://en.wikipedia.org/wiki/All-time_Olympic_Games_medal_table

## Lets see if each year is accurate

```python
Rio_Games = athletes.loc[athletes["Year"] == 2016]
Rio_Games.head()
```

#### Groupby Each Event

```python
Rio_Women = Rio_Games.loc[Rio_Games["Sex"] == "F"]
Rio_Men = Rio_Games.loc[Rio_Games["Sex"] == "M"]

Rio_Women_Sports = Rio_Women.groupby(by="Sport")
Rio_Men_Sports = Rio_Men.groupby(by="Sport")
```

```python
Rio_Women_Sports["Age", "Height", "Weight"].mean().round({"Age": 1, "Height": 0, "Weight":0})
```

```python
Rio_Men_Sports["Age", "Height", "Weight"].mean().round({"Age": 1, "Height": 0, "Weight":0})
```

## Now that we can definitely see a difference between athletes, lets plot line graphs to see a trend over time

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# For each sport, need a dataframe for mean (age, height, weight) by year
# Lets start with Gymnastics

Gymnastics = athletes.loc[athletes["Sport"] == "Gymnastics"]
```

```python
Women_Gymnastics = Gymnastics.loc[Gymnastics["Sex"] == "F"]
Men_Gymnastics = Gymnastics.loc[Gymnastics["Sex"] == "M"]

Year_Women_Gymnastics = Women_Gymnastics.groupby("Year", as_index=False)
Year_Men_Gymnastics = Men_Gymnastics.groupby("Year", as_index=False)
```

```python
Women_Gymnastics
```

```python
Year_Men_Gymnastics.mean()
```

```python
sns.lmplot(x="Year", y="Weight", data=Women_Gymnastics)
```

```python
fig, (ax1, ax2, ax3) = plt.subplots(figsize=(15,10), ncols=3)
sns.regplot(x="Year", y="Height", data=Year_Women_Gymnastics.mean(), ax=ax1)
sns.regplot(x="Year", y="Weight", data=Year_Women_Gymnastics.mean(), ax=ax2)
sns.regplot(x="Year", y="Age", data=Year_Women_Gymnastics.mean(), ax=ax3)

ax3.set_ylim(top=30)
```
