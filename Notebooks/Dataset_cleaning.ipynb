```python
import os
import numpy as np
import pandas as pd
```

```python
path_to_csv = os.path.join("..", "data", "athlete_events.csv")

athletes = pd.read_csv(path_to_csv)
```

```python
athletes.head()
```

```python
athletes.info()
```

```python
athletes_cleaned = athletes[np.isfinite(athletes["Age"])]
athletes_cleaned = athletes_cleaned[np.isfinite(athletes["Height"])]
athletes_cleaned = athletes_cleaned[np.isfinite(athletes["Weight"])]
athletes_cleaned.head()
```

```python
athletes_cleaned.info()
```

```python
athletes_cleaned.count()
```

### athletes_cleaned: now has NA values for weight, height, and age removed. 

```python
athletes_cleaned.to_csv(os.path.join("..", "data", "athlete_cleaned.csv"))
```

```python
country_group = athletes_cleaned.groupby("Team")
```
